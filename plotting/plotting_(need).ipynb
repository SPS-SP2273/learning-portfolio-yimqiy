## 1 A simple plot

### 1.1 Letâ€™s look at some code.

```python
import matplotlib.pyplot as plt
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, c='purple', ls='dashed', lw = 2, label='Y values')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

### 1.2 Adding another plot

```python
plt.plot(x, y_1, c='purple', ls='dashed', lw = 2, label='Y values')
plt.plot(x, y_2, color='blue', label='Y2 values')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
#everything about this plot resets until you set them again
```

### 1.3 Yet another plot but with error bars

```python
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

plt.plot(x, y_1, c='purple', ls='dashed', lw = 2, label='Y values')
plt.plot(x, y_2, color='blue', label='Y2 values')
plt.errorbar(x, y_3, lw = 1, xerr = x_error, yerr = y_error, 
             color = "green", label = "Y3 with errors")

plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

### 1.4 Experiment, Experiment, Experiment

```python
plt.plot(x, y_1, c='#d04153', ls='dashed', lw = 2, label='Y values')
plt.plot(x, y_2, color='#93d032', label='Y2 values')
plt.errorbar(x, y_3, lw = 1, xerr = x_error, yerr = y_error, 
             color = "#ff33ff", label = "Y3 with errors")

plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('Fun fact: few people care about these colours', alpha=0.4, c = 'blue')
plt.grid(alpha=.55, c = 'red', ls = 'dashed')
plt.legend(loc='lower right')
plt.show()
```

```python
plt.plot(x, y_1, c='#d04153', ls='dashed', lw = 2, label='Y values')
plt.plot(x, y_2, color='#93d032', label='Y2 values')
plt.errorbar(x, y_3, lw = 1, xerr = x_error, yerr = y_error, 
             color = "#ff33ff", label = "Y3 with errors")

plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=1) 
plt.legend() #supported values are 'best', 'upper right', 'upper left', 'lower left',
#'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'

plt.show()
```

### 1.5 Better with NumPy

```python
import numpy as np
np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
plt.show()
```

### 1.6 Adding mathematical functions

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend()
plt.show()
```

### 1.7 We need another axis!

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

# This creates a new y-axis for the plot that comes after
plt.twinx()
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')

plt.show()
```

### 1.8 Saving to disc

```python
plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

# This creates a new y-axis for the plot that comes after
plt.twinx()
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')
plt.show()
```

## 2 A real example: Global Warming

### 2.1 Plotting data from files

```python
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]
```

```python
plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
```

```python
plt.style.use('fivethirtyeight')

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
plt.show()
```

## Exercise 2 :  A simple mathematical plot

```python
# Some data for plotting
x = np.linspace(0,50,200)
y = np.exp(-x/10)*np.sin(x)

plt.plot(x, y)
plt.xlabel('x',fontsize=15, color='darkblue')
plt.ylabel('y',fontsize=15, color='darkblue')
plt.title(r'$y=e^{-x/10} \sin(x)$',size=20)
plt.ylim(-1,1)
plt.grid(alpha=.25)
plt.legend(loc = 'lower right')
#save_file_figure(plt, argv)                  # Hide
plt.show(block=False)
```
